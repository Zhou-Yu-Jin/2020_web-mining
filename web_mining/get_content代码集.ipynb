## 一、goal17

```python
import pandas as pd
import requests
from lxml import etree
import random # 随机模块
import time # import time：时间戳时间；结构化时间；格式化时间
```

```python
headers = {
   "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36"
}

#要改自己的
```

```python
def parse_url(url):
    response = requests.get(url,headers=headers)
    return response.content.decode()
```

```python
file_path = "目标URL.xlsx"

df = pd.read_excel(file_path)
url_list = df["url"].tolist()

```

```python
url_list
```

```python
content_list = []
```

```python
for url in url_list:
    html_str = parse_url(url)
    html = etree.HTML(html_str)
  #  time_random = random.randrange(5,30) # 随机翻页
   # time.sleep(time_random)
    item = {}
    item["title"] = html.xpath("//*[@id=\"wrapper\"]/div[3]/div/div/div/div/h1/text()")
    item["title"] = [i.replace("\n","").replace(" ","") for i in item['title']]
    item["title"] = ''.join(item["title"])

    item["content"] = html.xpath("//*[@class=\"fusion-text\"]/p/text()")
    item["content"] = [i.replace("\n","").replace(" ","").replace("▲","") for i in item['content']]
    item["content"] = ''.join(item["content"])
    content_list.append(item)
```

```python
name = ['title','content']
test = pd.DataFrame(columns=name,data=content_list)
test.to_csv("get_content.csv",mode='a',encoding='utf-8')
print("保存成功")
```
